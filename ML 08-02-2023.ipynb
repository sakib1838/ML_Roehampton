```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\sakiba\Downloads\bill_authentication.csv')
dataset.head(20)
```

```python
dataset.shape
```

```python
X = dataset.drop(['Class'], axis=1)
Y = dataset[['Class']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
disp.plot()

plt.show()
```

```python
activity_dataset = pd.read_csv(r'C:\Users\sakiba\Downloads\PastHires.csv')
activity_dataset.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

activity_dataset['Employed?']= le.fit_transform(activity_dataset['Employed?'])
activity_dataset
```

```python
activity_dataset['Level of Education']= le.fit_transform(activity_dataset['Level of Education'])
activity_dataset
```

```python
activity_dataset['Top-tier school']= le.fit_transform(activity_dataset['Top-tier school'])
activity_dataset
```

```python
activity_dataset['Interned']= le.fit_transform(activity_dataset['Interned'])
activity_dataset
```

```python
activity_dataset['Hired']= le.fit_transform(activity_dataset['Hired'])
activity_dataset
```

```python
AX = activity_dataset.drop(['Hired'],axis=1)
AY = activity_dataset[['Hired']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(AX, AY, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_clf = DecisionTreeClassifier()
DT_clf.fit(X_train,y_train)

DT_y_pred = DT_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, DT_y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_clf = RandomForestClassifier(max_depth=2)
RF_clf.fit(X_train,y_train)

RF_y_pred = RF_clf.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, RF_y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
GNB_clf = GaussianNB()
GNB_clf.fit(X_train,y_train)

GNB_y_pred = GNB_clf.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, RF_y_pred))
```
